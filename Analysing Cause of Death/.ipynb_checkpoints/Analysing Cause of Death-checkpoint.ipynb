# List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import matplotlib.pyplot as plt
```

```python
# View the plot 
%matplotlib inline
```

```python
#cause of deaths
cause = 'Chronic Diseases', 'Unintentional Injuries','Alzheimers', 'Influenza and Pneumonia', 'Sepsis', 'Others'
#percentile
percentile = [62, 5, 4, 2, 1, 26]
```

####  Draw a pie chart using the two data points

```python
#set the pie chart plot properties
#set figure size
plt.figure(figsize=(10,10))
```

```python
#explode the largest pie in the data set
explode = (0.05,0,0,0,0,0)
```

```python
#set pie chart properties
plt.pie(percentile,  labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)
plt.axis('equal')
plt.title('Causes of death in Ohio State')
plt.show()
```
